```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Salary_Data.csv')
x= df.iloc[ : , :-1].values
y= df.iloc[ : ,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_split, y_train, y_split = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_predict=lr.predict(X_split)
```

```python
plt.scatter(X_train,y_train, color='red')
plt.plot(X_train, lr.predict(X_train))
```

```python
plt.scatter(X_split,y_split, color='red')
plt.plot(X_train, lr.predict(X_train))
```
